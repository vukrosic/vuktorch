# Day 04: Tensor Ops

**Series:** PyTorch from Scratch: The Zero-to-One Framework

Day 04 upgrades tensor math with broadcasting and matrix multiply.

## Goal

Add the operations neural networks actually need: broadcasting, matmul, and reduction helpers.

## Big Idea

Broadcasting is just a shape rule. Matmul is the linear algebra core that powers dense layers.

## Build Here

Implement shape-aware gradients for add/mul and a clear backward rule for `@`.

## Exercise

Multiply a batch matrix by a weight matrix and verify the output shape and gradient flow.
