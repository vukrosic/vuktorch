# Day 05: Modules and Neurons

**Series:** PyTorch from Scratch: The Zero-to-One Framework

Day 05 starts the object model for reusable layers.

## Goal

Build a `Module` base class, a `Neuron`, and a clean way to collect parameters.

## Big Idea

A network is just a tree of reusable pieces that know how to expose their parameters.

## Build Here

Write the classes from scratch and make sure `parameters()` finds every weight and bias.

## Exercise

Create one neuron, run a forward pass, and print the parameter list.
