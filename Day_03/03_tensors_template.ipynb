# Day 03: Tensors

**Series:** PyTorch from Scratch: The Zero-to-One Framework

Day 03 is the first vectorized step. We move from scalars to arrays and keep the math readable.

## Goal

Build a NumPy-backed `Tensor` class that supports same-shape elementwise math and reductions.

## Big Idea

A tensor is still just data plus gradient history. The difference is that the data is now an array instead of a single number.

## Build Here

In code, define how tensors add, multiply, raise to a power, and reduce with `sum` and `mean`.

## Exercise

Create two same-shaped tensors, combine them, call `backward`, and inspect the gradients.
