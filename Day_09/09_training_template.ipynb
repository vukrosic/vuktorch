# Day 09: Training Loops

**Series:** PyTorch from Scratch: The Zero-to-One Framework

Day 09 assembles the pieces into a real training script.

## Goal

Write a clean training loop with forward pass, loss computation, backward pass, and updates.

## Big Idea

A training loop is the glue that repeats the forward, backward, and update steps until the model improves.

## Build Here

Implement one epoch, one evaluation pass, and a few lines of logging.

## Exercise

Train on a tiny synthetic dataset and print the loss over time.
