# Day 02: Autograd

**Series:** PyTorch from Scratch: The Zero-to-One Framework

Day 02 turns manual backprop into an actual `.backward()` method.

## Goal

Traverse the scalar graph in reverse topological order and call each local backward rule exactly once.

## Big Idea

Backprop is still the chain rule. The only new part is graph traversal.

## Build Here

Write `build_topo`, seed the output gradient with `1.0`, then walk the nodes in reverse order.

## Exercise

Use `.backward()` on `d = a * b + c` and confirm that the gradients match the manual Day 01 result.
