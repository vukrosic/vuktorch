# Day 08: Optimization

**Series:** PyTorch from Scratch: The Zero-to-One Framework

Day 08 adds the optimizer that turns gradients into learning.

## Goal

Build a simple `SGD` optimizer with `zero_grad` and a parameter update step.

## Big Idea

Gradient descent is just repeated small parameter nudges in the direction of lower loss.

## Build Here

Write the optimizer and show a single step on a toy example.

## Exercise

Run one gradient descent step and check that the loss decreases.
