# Day 06: Layers and MLPs

**Series:** PyTorch from Scratch: The Zero-to-One Framework

Day 06 composes neurons into layers and layers into a small MLP.

## Goal

Implement `Layer` and `MLP` so model composition feels natural.

## Big Idea

Once parameters are discoverable, scaling a model is mostly about composing smaller modules.

## Build Here

Write a `Layer` that holds neurons, then stack layers into an `MLP`.

## Exercise

Instantiate a tiny MLP and run one forward pass on toy input.
